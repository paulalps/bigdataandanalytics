```python
# Import Histogram, output_file, and show from bokeh.charts
from bokeh.charts import Histogram, output_notebook, show

# Import pandas as pd
import pandas as pd

# Import the fertility.csv data: data
data = pd.read_csv("imd_student_blind.csv", encoding = 'latin2')

# Make the Histogram: p
p = Histogram(data, 'ano_ingresso', title='Ano Ingresso', bins=20)

# Set axis labels
p.xaxis.axis_label = 'Ano ingresso (% population)'
p.yaxis.axis_label = 'Numero alunos'


# Call the output_notebook() 
output_notebook()
show(p)
```

```python
# Import Histogram, output_file, and show from bokeh.charts
from bokeh.charts import Histogram, output_notebook, show

# Import pandas as pd
import pandas as pd

# Import the fertility.csv data: data
data = pd.read_csv("imd_student_blind.csv", encoding = 'latin2')

# Make a Histogram: p
p = Histogram(data, '', title='Female Literacy',
              color='Continent', legend='top_left', bins=10)

# Set axis labels
p.xaxis.axis_label = 'Female Literacy (% population)'
p.yaxis.axis_label = 'Number of Countries'


# Call the output_notebook() 
output_notebook()
show(p)
```

```python
# Import Scatter, output_notebook, and show from bokeh.charts
from bokeh.charts import Scatter, output_notebook, show

# Import pandas as pd
import pandas as pd

# Import the fertility.csv data: data
data = pd.read_csv("imd_student_blind.csv", encoding = 'latin2')


# Make a scatter plot: p
p = Scatter(data, x='CEP', y='nota',
            title='CEP vs notas')

# Set the x-axis label
p.xaxis.axis_label = 'CEP'

# Set the y-axis label
p.yaxis.axis_label = 'nota'

# Call the output_notebook() 
output_notebook()
show(p)
```
